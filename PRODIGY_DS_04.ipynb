```
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Load the CSV file
file_path = 'D:\\twitter_training.csv'
df = pd.read_csv(file_path, names=['ID', 'Entity', 'Sentiment', 'Content'])

# Display the first few rows of the DataFrame
df.head()
```

```
# DataFrame details
print("DataFrame Shape:", df.shape)
print("DataFrame Description:")
print(df.describe())
print("Missing Values Before Dropping:")
print(df.isnull().sum())
```

```
# Drop rows with missing values
df.dropna(axis=0, inplace=True)

# Check for missing values again
print("Missing Values After Dropping:")
print(df.isnull().sum())
```

```
# Check for duplicate rows
print("Duplicate Rows Count Before Dropping:", df.duplicated().sum())

# Drop duplicate rows
df.drop_duplicates(inplace=True)

# Check for duplicate rows again
print("Duplicate Rows Count After Dropping:", df.duplicated().sum())
print("DataFrame Shape After Cleaning:", df.shape)
```

```
# Plot sentiment distribution with customized colors
sentiment_counts = df['Sentiment'].value_counts()
plt.figure(figsize=(6, 3))
sentiment_counts.plot(kind='bar', color=['#FF6347', '#4682B4', '#32CD32', '#FFD700'])  # Customized colors
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.xticks(rotation=0)
plt.show()
```

```
# Filter data for the entity 'Microsoft'
brand_data = df[df['Entity'].str.contains('Microsoft', case=False)]
brand_sentiment_counts = brand_data['Sentiment'].value_counts()
```

```
# Plot sentiment distribution for Microsoft
plt.figure(figsize=(6, 6))
plt.pie(brand_sentiment_counts, labels=brand_sentiment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Sentiment Distribution for Microsoft')
plt.show()
```
